```python
import os
import json

# 指定資料夾路徑
folder_path = r"c:\Users\George\Desktop\專題Program\專題\Dataset\agent-as-a-judge\benchmark\devai\instances"

# 遍歷資料夾中的所有 JSON 檔案
for file_name in os.listdir(folder_path):
    if file_name.endswith(".json"):  # 確保只處理 JSON 檔案
        json_file_path = os.path.join(folder_path, file_name)
        
        # 讀取 JSON 檔案
        with open(json_file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
        
        # 列出所需的欄位
        print("Name:", data.get("name", "N/A"))
        print("Query:", data.get("query", "N/A"))
        print("-" * 50)
```
